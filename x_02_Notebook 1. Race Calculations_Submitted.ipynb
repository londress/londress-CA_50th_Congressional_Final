### Creating Other Racial Category:

Aggregating the racial categories in order to analyze only Black, Asian, Latino, and White groups. The following will be considered "Other agg variable":  

    -American Indian and Alaska Native alone 
    
    -Native Hawaiian and Other Pacific Islander
    
    -Some other race alone

    -Two or more races


Variable names:     HD01_VD05, HD01_VD07, HD01_VD08, HD01_VD09


```python
# manually calculating estimates for other racial categories
df["Other_agg"] = (df["HD01_VD05"] + 
                                        df["HD01_VD07"] + 
                                        df["HD01_VD08"] + 
                                        df["HD01_VD09"])
```

###### Creating Margin of Error Variables

```python
# sum of square for MOEs
def calc_est(est):
    est_sum = np.sum(est) # sum all estimates
    return est_sum
```

```python
# selcting columns by position, df_m.iloc[:, [7, 9, 10, 11]]
df_m['Other_Agg_MOE'] = df_m.iloc[:, [7, 9, 10, 11]].apply(calc_est, axis=1)
df_m.loc[:,'Other_Agg_MOE':].head(3)
```

```python
# renaming columns
df_m.rename(columns={"HD02_VD01" : "Total",
                    "HD02_VD02" : "Not_Latino",
                    "HD02_VD03" : "White",
                    "HD02_VD04" : "Black",
                    "HD02_VD05" : "Native",
                    "HD02_VD06" : "Asian",
                    "HD02_VD07" : "API",
                    "HD02_VD08" : "Other",
                    "HD02_VD09" : "Two_plus",
                    "HD02_VD12" : "Latino"}, inplace=True)
```

```python
# creating a table with MOES only. 
df_MOE = df_m
```

```python
# Renaming column names with a suffix
df_MOE = df_MOE.add_suffix('_MOE')
```

```python
df_MOE.head(1)
```

```python
df_MOE = df_MOE.rename(columns = {'Other_Agg_MOE_MOE' : 'Other_Agg_MOE'})
df_MOE.head(2)
```

```python
df_MOE_pct = df_MOE.loc[:,"Not_Latino_MOE":"Other_Agg_MOE"].div(df_MOE["Total_MOE"], axis=0)
df_MOE_pct = df_MOE_pct.add_suffix('_pct')
df_MOE_pct
```

```python
df3 = df_MOE.join(df_MOE_pct, how='inner')
df3.head(1)
```

```python
df3.drop(df3.columns[[0, 1, 2]], axis = 1, inplace = True)
df3.head(1)
```

```python
df_final = df2.join(df3, how='inner')
df_final
```
